```python
# Övning 5A,
import random

roll_list = []
six_list = []

for i in range(100):
    roll_list.append(random.randint(1, 6))

for x in roll_list:
    if x == 6:
        six_list.append(6)

num_of_sixes = len(six_list)

print(f"{num_of_sixes} sixes was rolled out of 100 rolls.")
```

```python
# Övning 5B, 
random.seed(1)


number_of_sixes = []
probability = []
number_of_rolls = [10, 100, 1000, 10000, 100000, 1000000]

for roll in number_of_rolls:
    count = 0
    for i in range(roll):
        if random.randint(1, 6) == 6:
            count += 1

    number_of_sixes.append(count)
    probability.append(count/roll)

for rolls, sixes, probab in zip(number_of_rolls, number_of_sixes, probability):
    print(f"{rolls, sixes, probab}")
```

```python
# Övning 5C,

import random
import matplotlib.pyplot as plt
import numpy as np


plt.plot(probability, '-*')
plt.title("Probability of sixes in increasing number of rolls")
plt.xticks([0,1,2,3,4,5], number_of_rolls)
plt.xlabel("Number of dice rolls")
plt.ylabel("Probability")

plt.show
```
