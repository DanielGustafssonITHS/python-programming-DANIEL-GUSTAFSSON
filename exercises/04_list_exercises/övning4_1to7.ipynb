```python
# Övning 5A,
import random

roll_list = []
six_list = []

for i in range(100):
    roll_list.append(random.randint(1, 6))

for x in roll_list:
    if x == 6:
        six_list.append(6)

num_of_sixes = len(six_list)

print(f"{num_of_sixes} sixes was rolled out of 100 rolls.")
```

```python
# Övning 5B, 
random.seed(1)


number_of_sixes = []
probability = []
number_of_rolls = [10, 100, 1000, 10000, 100000, 1000000]

for roll in number_of_rolls:
    count = 0
    for i in range(roll):
        if random.randint(1, 6) == 6:
            count += 1

    number_of_sixes.append(count)
    probability.append(count/roll)

for rolls, sixes, probab in zip(number_of_rolls, number_of_sixes, probability):
    print(f"{rolls, sixes, probab}")
```

```python
''' 
This is called the Monty Hall problem and is a "famous" probability conundrum and I'm not sure why we dont call it that...

The best choice according to probability is to always switch as at you first choice odds of you being correct was 1/3. When the other door opens which we know is
one of the wrong answers the probability changes so that the next door has the odds of both doors which means that your door still has the probability of 1/3 and
the other door thats left now has the probability of 2/3 so 2 out of 3 times it will be correct to change making it the best choice.
'''
import random

correct_door = random.randint(1, 3)

first_choice = input("Pick a door between 1 and 3: ")


```

```python
import random
import matplotlib.pyplot as plt

def monty_hall(switch: bool, trials: int) -> float:
    wins = 0
    for _ in range(trials):
        # Slumpa vilken dörr som har kaninen (vinsten)
        rabbit = random.randint(0, 2)
        # Spelaren väljer en dörr
        choice = random.randint(0, 2)
        # Monty öppnar en dörr utan kanin
        possible_doors = [d for d in range(3) if d != rabbit and d != choice]
        monty_opens = random.choice(possible_doors)
        # Om spelaren byter
        if switch:
            remaining = [d for d in range(3) if d not in (choice, monty_opens)][0]
            choice = remaining
        # Kontrollera vinst
        if choice == rabbit:
            wins += 1
    return wins / trials

# Antal simuleringar
trials_list = [10, 100, 1000, 10000, 100000, 1000000]

stay_results = []
switch_results = []

for n in trials_list:
    stay_results.append(monty_hall(False, n))
    switch_results.append(monty_hall(True, n))

# Plotta resultaten
plt.figure(figsize=(10,6))
plt.plot(trials_list, stay_results, marker='o', label="Stanna kvar")
plt.plot(trials_list, switch_results, marker='o', label="Byta dörr")
plt.axhline(1/3, color='red', linestyle='--', label="Teoretiskt: stanna (1/3)")
plt.axhline(2/3, color='green', linestyle='--', label="Teoretiskt: byta (2/3)")
plt.xscale('log')  # log-skala gör det lättare att se
plt.xlabel("Antal simuleringar (log-skala)")
plt.ylabel("Andel vinster")
plt.title("Monty Hall-simulering")
plt.legend()
plt.show()
```
