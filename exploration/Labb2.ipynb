```python
# Läsa in  datan i lämplig datastruktur
import matplotlib.pyplot as plt
import numpy as np
import math
import random


datapoints = []
path_datapoints = "/mnt/c/Users/danie/OneDrive/Skrivbord/labb2/datapoints.txt"
testpoints = []
path_testpoints = "/mnt/c/Users/danie/OneDrive/Skrivbord/labb2/testpoints3.txt"

width_dp = [] 
height_dp = []
pokemon = []

width_tp = [] 
height_tp = []
pokemontp = []


with open(path_datapoints,) as lines:
    next(lines)
    for line in lines:
        row = line.split(",")
        width_dp.append([float(row[0])])
        height_dp.append([float(row[1])])
        pokemon.append([float(row[2])])


with open(path_testpoints,) as lines:
    
    for line in lines:
        row = line.strip().split(",")
        width_tp.append([float(row[0])])
        height_tp.append([float(row[1])])
        pokemontp.append([float(row[2])])        
#print(width_dp)
#print(height_dp)
#print(pokemon)
width_dp = np.array(width_dp)
height_dp = np.array(height_dp)
width_tp = np.array(width_tp)
height_tp = np.array(height_tp)

plt.title("Length & Girth")
plt.xlabel("Width")
plt.ylabel("Height")    
#plt.plot(width_dp, height_dp, "o", c = "hotpink") 
plt.scatter(width_dp, height_dp, c='green', label='dp', alpha=0.7)   
plt.scatter(width_tp, height_tp, c='blue', label='tp', alpha=0.7)
plt.legend()
plt.show()
```
